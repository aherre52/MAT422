#   2.2.1. Probability axioms

An experiment in probability refers to any process with uncertain outcomes, including activities like tossing a coin or measuring blood pressure. The sample space, denoted by $S$, represents all possible outcomes of the experiment. An event is a subset of outcomes within the sample space, where simple events consist of a single outcome and compound events consist of multiple outcomes.

The probability distribution assigns a probability $P(A)$ to each event $A$, adhering to the following axioms

* For any event $A$, $0 \leq P(A) \leq 1$.
* $P(S) = 1$.
* If $A_1, A_2, A_3, \ldots$ is an infinite collection of disjoint events, then $P(A_1 \cup A_2 \cup A_3 \cup \cdots) = \sum_{i=1}^\infty P(A_i)$.
* For any event $A$, $P(A) + P(A^c) = 1$, from which $P(A) = 1 - P(A^c)$.
* When events $A$ and $B$ are mutually exclusive, $P(A \cup B) = P(A) + P(B)$.
* For any two events $A$ and $B$, $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.


```python
#will use itertools to make a deck of cards
import itertools

# Define suits and suits for a deck of cards
suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']

# Create the sample space of a deck of 52 cards)
deck_of_cards = list(itertools.product(ranks, suits))
print(f"Sample space (Deck of cards): {deck_of_cards}\n")

# Will now define events
# Event A: The card is an odd number (excluding ace and jack, queen, king)
odd_numbers = ['3', '5', '7', '9']
event_A = [(rank, suit) for rank, suit in deck_of_cards if rank in odd_numbers]
print(f"Event A (Card is an odd number): {event_A}")

# Event B: The card is a Heart
event_B = [(rank, suit) for rank, suit in deck_of_cards if suit == 'Hearts']
print(f"Event B (Card is a heart): {event_B}")

# Define a probability function for equally likely outcomes
def probability(event, sample_space):
    #based off of event cardinality and sample space cardinality
    return len(event) / len(sample_space)

# Calculate probabilities
P_A = probability(event_A, deck_of_cards)
P_B = probability(event_B, deck_of_cards)

print(f"\nP(A) = Probability of event A (picking an odd numbered card): {P_A:.2f}")
print(f"P(B) = Probability of event B (picking a card with suit of hearts): {P_B:.2f}")

# Demonstrating probability axioms mentioned in chapter
# 1. P(A ∩ B): Probability of card being both an odd number and a heart
event_A_intersect_B = [(rank, suit) for rank, suit in event_A if suit == 'Hearts']
P_A_intersect_B = probability(event_A_intersect_B, deck_of_cards)
print(f"P(A ∩ B) = Probability of event A ∩ B (Odd number and Heart): {P_A_intersect_B:.2f}")

# 2. P(A ∪ B) = P(A) + P(B) - P(A ∩ B)
P_A_union_B = P_A + P_B - P_A_intersect_B
print(f"P(A ∪ B) = Probability of event A ∪ B (Odd number or Heart): {P_A_union_B:.2f}")

# 3. P(A') = 1 - P(A), where A' is the complement of A (not an odd number)
P_A_complement = 1 - P_A
print(f"P(A') = Probability of event A' (Not an odd number): {P_A_complement:.2f}")

# Assign equal probabilities to all outcomes, as they are each likely
N = len(deck_of_cards)
print(f"\nEach card has an equal probability: 1/{N} = {1/N:.4f}")
```

#    2.2.2. Conditional probability

Conditional probability measures the likelihood of an event $A$ occurring given that another event $B$ has occurred. It is defined as $P(A \mid B) = \frac{P(A \cap B)}{P(B)}$, where $P(A \cap B)$ is the probability of both events occurring and $P(B)$ is the probability of the conditioning event $B$. This concept leads to the multiplication rule $P(A \cap B) = P(A \mid B) \cdot P(B)$, which is useful when calculating joint probabilities. Additionally, events $A$ and $B$ are independent if $P(A \mid B) = P(A)$, and this independence can be extended to collections of more than two events.





```python
# Will use the same deck of cards example for conditional probability

# Define suits and values for a deck of cards
suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
ranks = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']

# Create the sample space of a deck of cards (52 cards)
deck_of_cards = list(itertools.product(ranks, suits))

# Event A: The card is an Ace
event_A = [(rank, suit) for rank, suit in deck_of_cards if rank == 'Ace']

# Event B: The card is a Heart
event_B = [(rank, suit) for rank, suit in deck_of_cards if suit == 'Hearts']

# Intersection of A and B: The card is an Ace of Hearts
event_A_intersect_B = [(rank, suit) for rank, suit in event_A if suit == 'Hearts']

# Define a probability function for equally likely outcomes
def probability(event, sample_space):
    return len(event) / len(sample_space)

# Calculate unconditional probabilities
P_A = probability(event_A, deck_of_cards)
P_B = probability(event_B, deck_of_cards)
P_A_intersect_B = probability(event_A_intersect_B, deck_of_cards)

# Calculate conditional probability P(A | B) = P(A ∩ B) / P(B)

P_A_given_B = P_A_intersect_B / P_B


print(f"Total number of cards: {len(deck_of_cards)}")
print(f"Event A (Ace): {event_A}")
print(f"Event B (Hearts): {event_B}")
print(f"Event A ∩ B (Ace of Hearts): {event_A_intersect_B}\n")

print(f"P(A) = Probability of drawing an Ace: {P_A:.2f}")
print(f"P(B) = Probability of drawing a Heart: {P_B:.2f}")
print(f"P(A ∩ B) = Probability of drawing the Ace of Hearts: {P_A_intersect_B:.2f}")
print(f"P(A | B) = Conditional probability of Ace given the card is a Heart: {P_A_given_B:.2f}")

# Conditional Probability (Multiplication Rule)
# P(A ∩ B) = P(A | B) * P(B)
print(f"\nMultiplication Rule Check: P(A ∩ B) = P(A | B) * P(B) = {P_A_given_B * P_B:.2f}")
```

#  2.2.3. Discrete random variables

A random variable is a function mapping outcomes of an experiment to real numbers, with discrete random variables having either a finite set of values or an infinite sequence of values. The probability mass function (pmf) $p(x) = P(X = x)$ defines the probability of a discrete random variable taking a specific value, while the cumulative distribution function (cdf) $F(x) = P(X \leq x)$ accumulates these probabilities. The expected value of a discrete random variable $X$ is $$E(X) = \sum_{x \in D} x \cdot p(x)$$ and the variance is $$V(X) = \sum_{x \in D} (x - \mu)^2 \cdot p(x)$$ where $\mu$ is the mean of $X$.

For a Bernoulli random variable with probability $p$, the expected value is $p$, and the pmf is:

$$
b(x; n, p) =
\begin{cases}
\binom{n}{x} p^x (1 - p)^{n - x} & \text{for } x = 0, 1, 2, \ldots, n \\
0 & \text{otherwise}
\end{cases}
$$

For a Poisson distribution with parameter $\mu$, both the mean and variance are $\mu$, and the pmf is:

$$
p(x; \mu) = \frac{e^{-\mu} \mu^x}{x!}, \quad x = 0, 1, 2, 3, \ldots
$$


You are a florist who bought a seed packet containing 40 seeds. The likelihood of the flowers you plant having a genetic mutation that gives it red stripes instead of purple stripes is 0.35. Of the 40 seeds planted:
* (a) What is the probability that exactly 8 have the mutation?
* (b) What is the probability that 15 or less have the mutation?

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

# Parameters
n = 40  # number of seeds
p = 0.35  # probability of mutation (red stripes)

# Binomial distribution
def binomial_pmf(k, n, p):
    #implementation of the binomial probability mass function.
    return comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

def binomial_cdf(k, n, p):
    #implementation of the binomial cumulative distribution function
    cumulative_prob = 0
    for i in range(k + 1):
        cumulative_prob += binomial_pmf(i, n, p)
    return cumulative_prob

# Expected value (mean) and variance
def expected_value(n, p):
    # Calculate the expected value (mean) of the binomial distribution
    return n * p

def variance(n, p):
    # Calculate the variance of the binomial distribution.
    return n * p * (1 - p)

# Number of successful mutations observed
k = 8

# Calculate the probabilities
prob_k = binomial_pmf(k, n, p)

print(f"Probability of exactly {k} flowers with the mutation: {prob_k:.4f}")

k=15

cumulative_prob_k = binomial_cdf(k, n, p)

# Calculate mean and variance
mean = expected_value(n, p)
var = variance(n, p)

print(f"Cumulative probability of up to {k} mutations: {cumulative_prob_k:.4f}")
print(f"Expected value (mean): {mean:.2f}")
print(f"Variance: {var:.2f}")

# Plotting the binomial distribution
x = np.arange(0, n+1)
pmf_values = [binomial_pmf(i, n, p) for i in x]
cdf_values = [binomial_cdf(i, n, p) for i in x]

# Plot PDF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.title('Binomial Distribution PMF')
plt.xlabel('Number of Red Stripe Mutations')
plt.ylabel('Probability')
plt.grid(True)


# Plot CDF
plt.subplot(1, 2, 2)
plt.bar(x, cdf_values, color='green', alpha=0.7)
plt.title('Binomial Distribution CDF')
plt.xlabel('Number of Red Stripe Mutations')
plt.ylabel('Cumulative Probability')
plt.grid(True)

plt.tight_layout()
plt.show()
```

The number of phone orders to a small business pizza place can be modeled as a Poisson random variable. Assume that on the average there are 8 calls per hour.

* (a) What is the probability that there are exactly 5 calls in one
hour?

* (b) What is the probability that there are 8 or fewer calls in
one hour?

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 8  # average number of calls per hour

# Poisson PDF
def poisson_pmf(x, mu):
    #implementation of the Poisson probability mass function.
    return (math.exp(-mu) * (mu ** x)) / math.factorial(x)

# Poisson CDF
def poisson_cdf(x, mu):
    #Manual implementation of the Poisson cumulative distribution function
    cumulative_prob = 0
    for k in range(x + 1):
        cumulative_prob += poisson_pmf(k, mu)
    return cumulative_prob

# (a) Probability of exactly 5 calls in one hour
fiveCalls = 5
prob_exact = poisson_pmf(fiveCalls, mu)

# (b) Probability of 3 or fewer calls in one hour
maxEight = 8
prob_less_equal = poisson_cdf(maxEight, mu)

print(f"Probability of exactly {fiveCalls} calls in one hour: {prob_exact:.4f}")
print(f"Probability of 8 or fewer calls in one hour: {prob_less_equal:.4f}")

# Plotting Poisson PDF and CDF
x_range = np.arange(0, 20)
pdf_values = [poisson_pmf(x, mu) for x in x_range]
cdf_values = [poisson_cdf(x, mu) for x in x_range]

# Plot PDF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x_range, pdf_values, color='blue', alpha=0.7)
plt.title('Poisson Distribution PMF')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.ylim(0, max(pdf_values) * 1.1)  # Set y-axis to start at 0
plt.grid(True)

# Plot CDF
plt.subplot(1, 2, 2)
plt.bar(x_range, cdf_values, color='green', alpha=0.7)
plt.title('Poisson Distribution CDF')
plt.xlabel('Number of Calls')
plt.ylabel('Cumulative Probability')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# 2.2.4. Continuous random variables

A continuous random variable $X$ has possible values within an interval or a union of disjoint intervals on the number line. The probability density function (pdf) $f(x)$ of $X$ satisfies $f(x) \geq 0$ and the total area under the pdf is 1:
$$
\int_{-\infty}^{\infty} f(x) \, dx = 1.
$$
The probability that $X$ falls within the interval $[a, b]$ is given by:
$$
P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx.
$$
The expected value (mean) of $X$ is:
$$
\mu_X = E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx,
$$
and the variance is:
$$
\sigma_X^2 = V(X) = \int_{-\infty}^{\infty} (x - \mu_X)^2 \cdot f(x) \, dx.
$$
For an exponential distribution with rate $\lambda$, the pdf is $f(x; \lambda) = \lambda e^{-\lambda x}$ for $x \geq 0$, with expected value $\frac{1}{\lambda}$ and variance $\frac{1}{\lambda^2}$.

Normal distributions are widely used in various fields to model data that clusters around a central value, such as human heights or income levels. This distribution's familiar bell-shaped curve helps in analyzing and predicting real-world phenomena where data tends to follow a symmetrical pattern.
If $X$ has a normal distribution with mean $\mu$ and standard deviation $\sigma$, then
$$
Z = \frac{X - \mu}{\sigma}.
$$



The time between arrivals of taxis at a busy intersection is exponentially distributed with a mean of 10 minutes.

* (a) What is the standard deviation?

* (b) What is the probability that you wait longer than 20 minutes for a taxi?

* (c) What is the probability that you wait less than than 12 minutes for a taxi?


```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean_time = 10  # mean time in minutes
lambda_ = 1 / mean_time  # rate parameter

# Exponential PDF
def exponential_pdf(x, lambda_):
    #Probability density function of the exponential distribution.
    return lambda_ * np.exp(-lambda_ * x)

# Exponential CDF
def exponential_cdf(x, lambda_):
    #Cumulative distribution function of the exponential distribution.
    return 1 - np.exp(-lambda_ * x)

# (a) Find standard deviation

stdDev = math.sqrt(1/(lambda_**2))

# (b) Probability of waiting longer than 20 minutes
time_waited = 20
prob_longer_than_20mins = 1-exponential_cdf(20, lambda_)

# (c) Probability of waiting less than 12 minutes
time_waited = 12
prob_less_than_12mins = exponential_cdf(12, lambda_)

print(f"Standard deviation: {stdDev:.1f} minutes")
print(f"Probability of waiting longer than 20 minutes: {prob_longer_than_20mins:.4f}")
print(f"Probability of waiting less than 12 minutes: {prob_less_than_12mins:.4f}")

# Now want to plot the Exponential Distribution
x = np.linspace(0, 120, 1000)  # Range of time values for plotting
pdf_values = exponential_pdf(x, lambda_)  # PDF values

plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, color='blue')
plt.title('Exponential Distribution PDF')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.ylim(0, max(pdf_values) * 1.1)  # Set y-axis to start from 0
plt.grid(True)

# Plot CDF
cdf_values = [exponential_cdf(t, lambda_) for t in x]
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, color='green')
plt.title('Exponential Distribution CDF')
plt.xlabel('Time (minutes)')
plt.ylabel('Cumulative Probability')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Cholesterol is a fatty substance that is an important part
of the outer lining (membrane) of cells in the body of animals.
Its normal range for an adult is 120–240 mg/dl. The Food and
Nutrition Institute of the Philippines found that the total cholesterol level for Filipino adults has a mean of 159.2 mg/dl and
84.1% of adults have a cholesterol level less than 200 mg/dl
(http://www.fnri.dost.gov.ph/). Suppose that the total cholesterol level is normally distributed.
(a) Determine the standard deviation of this distribution.
(b) What are the quartiles (the 25% and 75% percentiles) of
this distribution?
(c) What is the value of the cholesterol level that exceeds 90%
of the population?

```python
# will use a stats library to avoid manual implementation
# normal distribution is widely used
import scipy.stats as stats

# Data provided from problem description
mean = 159.2  # Mean cholesterol level in mg/dl
percentile_84_1 = 200  # 84.1th percentile

# (a) Determine the standard deviation of this distribution
# Find the z-score for the 84.1th percentile
z_84_1 = stats.norm.ppf(0.841)  # use the percentile function to find z-score

# Calculate standard deviation using formula
std_dev = (percentile_84_1 - mean) / z_84_1

# (b) Calculate the quartiles (25% and 75% percentiles)
z_25 = stats.norm.ppf(0.25)  # Z-score for 25th percentile
z_75 = stats.norm.ppf(0.75)  # Z-score for 75th percentile

# Calculate quartile values
q1 = mean + z_25 * std_dev
q3 = mean + z_75 * std_dev

# (c) Find the cholesterol level that exceeds 90% of the population
percentile_90 = stats.norm.ppf(0.90)  # Z-score for 90th percentile
cholesterol_90 = mean + percentile_90 * std_dev

print(f"Standard Deviation: {std_dev:.2f} mg/dl")
print(f"25th Percentile (Q1): {q1:.2f} mg/dl")
print(f"75th Percentile (Q3): {q3:.2f} mg/dl")
print(f"Cholesterol level exceeding 90% of the population: {cholesterol_90:.2f} mg/dl")
```

# 3.2.1. Limits and continuity

Limits are fundamental in calculus and are used to define continuity, derivatives, and integrals. For a real-valued function $f: D \to \mathbb{R}$, the limit $L$ exists as $x$ approaches $a$ if for any $\epsilon > 0$, there exists a $\delta > 0$ such that $|f(x) - L| < \epsilon$ for all $x \in D \cap B_\delta(a) \setminus \{a\}$.

A function $f$ is continuous at a point $a$ if $\lim_{x \to a} f(x) = f(a)$. The composition of continuous functions $g \circ f$ is also continuous at the given point. Additionally, the extreme value theorem asserts that a continuous function on a closed and bounded set $D \subseteq \mathbb{R}^d$ attains both maximum and minimum values.


# Code description
The code visualizes the function $f(x) = \sin \frac{1}{x}$ near $x = 0$ to show the concept of limits. The function oscillates rapidly as $x$ approaches 0, demonstrating that the limit does not exist at $x = 0$.

The next part of the code plots a simple continuous function, $f(x) = x^2$ and a discontinous function, $f(x) = \frac{1}{x-1}$. This shows that as $x$ approaches any point, the function value smoothly follows, illustrating the definition of continuity, where $\lim_{x \to a} f(x) = f(a)$ for all $a$ in the domain.

Next in the code is an example demonstrating composition by defining $f(x) = x^2$ and $g(x) = \sin(x)$. It plots the composition $h(x) = g(f(x))$, showing that the composition of continuous functions remains continuous.

Lastly, the code computes and plots a function $f(x) = -x^2 + 4$ to illustrate extremum. It identifies the maximum and minimum values of $f(x)$, verifying that a continuous function on a closed interval attains a maximum and a minimum, in line with the Extreme Value Theorem.



```python
import numpy as np
import matplotlib.pyplot as plt

# Will demonstrate concept of limits with f(x) = sin(1/x) as x approaches 0
# have seen an example like this in MAT 371
def f_limit(x):
    return np.sin(1/x)

# getting an assortment of x values, to show rapid oscillation
x_values = np.linspace(0.01, 0.1, 400)
y_values = f_limit(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$f(x) = \sin(\frac{1}{x})$', color='blue')
plt.title("Behavior of Limit as x Approaches 0")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

# Will demonstrate concept of continuity with two different functions:
# f(x) = x^2 and f(x) = 1/(x - 1) (which is discontinuous at x=1)
def f_continuous(x):
    return x**2

def f_discontinuous(x):
    return 1 / (x - 1)

x_values_cont = np.linspace(-2, 2, 400)
y_values_cont = f_continuous(x_values_cont)

x_values_disc = np.linspace(0.1, 1.9, 400)  # Avoid x = 1 for discontinuity
y_values_disc = f_discontinuous(x_values_disc)

plt.figure(figsize=(10, 6))

# plotting continuous function
plt.subplot(1, 2, 1)
plt.plot(x_values_cont, y_values_cont, label=r'$f(x) = x^2$', color='green')
plt.title("Continuous Function")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# plotting discontinuous function, side by side
plt.subplot(1, 2, 2)
plt.plot(x_values_disc, y_values_disc, label=r'$f(x) = \frac{1}{x-1}$', color='red')
plt.title("Discontinuous Function")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Will do composition of functions with two continuous functions,
# f(x) = x^2 and g(x) = sin(x), and then do h(x) = g(f(x)), also continuous
def f(x):
    return x**2

def g(x):
    return np.sin(x)

def h(x):
    return g(f(x))

x_values = np.linspace(-2, 2, 400)
y_values_h = h(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_h, label=r'$h(x) = \sin(x^2)$', color='purple')
plt.title("Composition of Functions: h(x) = sin(x^2)")
plt.xlabel('x')
plt.ylabel('h(x)')
plt.grid(True)
plt.legend()
plt.show()

# Finally will do an extremum demonstration with f(x) = -x^2 + 4
# mainly, this fucntion has a maximum as its extremum
def f_extremum(x):
    return -x**2 + 4

x_values = np.linspace(-3, 3, 400)
y_values_extremum = f_extremum(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_extremum, label=r'$f(x) = -x^2 + 4$', color='orange')
plt.title("Extremum: Maximum of a Function")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(y=4, color='gray', linestyle='--')
plt.legend()
plt.show()
```

#    3.2.2. Derivatives


Derivatives are a fundamental concept in calculus that represent the rate of change of a function. For a real-valued function $f: D \to \mathbb{R}$, the derivative at a point $a$, denoted $f'(a)$, exists if the limit

$$
f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}
$$

exists. A function is said to be differentiable at a point if this limit exists, which implies that the function is continuous at that point. Moreover, if $f$ and $g$ are differentiable functions, the derivative of their linear combination, $\alpha f(x) + \beta g(x)$, is given by

$$
(\alpha f + \beta g)'(x) = \alpha f'(x) + \beta g'(x),
$$

where $\alpha$ and $\beta$ are constants. Additionally, the application of differentiation rules allows for the analysis of polynomial and trigonometric functions, providing insight into their behavior.

Partial derivatives measure how a function changes as one variable changes while keeping other variables constant. For a function $ f(x_1, x_2, \ldots, x_n) $, the partial derivative with respect to $ x_i $ is denoted as $ \frac{\partial f}{\partial x_i} $.

The chain rule provides a method for computing the derivative of a composite function. If $ y = f(g(x)) $, then the chain rule states that
$$
\frac{dy}{dx} = \frac{df}{dg} \cdot \frac{dg}{dx}.
$$

The Hessian matrix, denoted as $ H $, consists of all second-order partial derivatives of a scalar-valued function. It is instrumental in optimization, as it provides information about the local curvature of the function. A positive definite Hessian indicates a local minimum, while a negative definite Hessian suggests a local maximum, thus aiding in efficiently finding critical points.


Overall, derivatives play a crucial role in understanding the dynamics of functions and their applications in various fields.



# Derivative code demonstration

The code begins by defining a single-variable function $f(x) = x^2 + 3x + 2$. It then calculates the derivative $f'(x)$.

Next, the code demonstrates the linearity of derivatives by combining two functions, $f(x)$ and $g(x) = 2x^3 - x + 1$, with constants $\alpha$ and $\beta$. It computes the derivative of the combined function $\alpha f(x) + \beta g(x)$, reinforcing the principle that the derivative of a sum is the sum of the derivatives.

The Mean Value Theorem is illustrated by calculating the average rate of change of the function $f(x) = x^2$ over the interval $[a, b] = [1, 3]$. The code verifies that there exists a point $c$ in the interval such that $f'(c)$ equals this average rate of change, showcasing the theorem's application.

Finally, the code computes the Hessian matrix for the function $f(x, y) = x^2 + y^2$. The Hessian matrix, which consists of the second-order partial derivatives, provides insights into the curvature of the function, aiding in optimization problems.




```python
# using the sympy library instead of numpy for symbolic math
import sympy as sp

# Will first do a single-variable derivative
# First, will define a single-variable function f(x)
x = sp.symbols('x')
f = x**2 + 3*x + 2  # A quadratic function

# Calculate the derivative of f with respect to x
f_prime = sp.diff(f, x)
print(f"f(x) = x^2 + 3x + 2")
print(f"\nDerivative of f(x) is f'(x) = {f_prime}")

# Demonstrate linearity of derivatives
# Define g(x), use same f(x), and constants alpha, beta
g = 2*x**3 - x + 1  # Another function
alpha, beta = sp.symbols('alpha beta')

# Linearity of derivatives: [alpha*f(x) + beta*g(x)]'
combined_function = alpha * f + beta * g
combined_derivative = sp.diff(combined_function, x)

print(f"\ng(x) = 2x^3 - x + 1")

print(f"\nDerivative of alpha*f(x) + beta*g(x) is: {combined_derivative}")

# Demonstrate Mean Value Theorem
# Define the function f(x) = x^2 on [a, b]
a, b = 1, 3  # Interval
f_mvt = x**2

# The Mean Value Theorem guarantees there is some c in (a, b)
# such that f(b) - f(a) / (b - a) = f'(c)
# Let's calculate the average rate of change
average_rate_of_change = (f_mvt.subs(x, b) - f_mvt.subs(x, a)) / (b - a)

# Now solve f'(c) = average_rate_of_change for c
f_mvt_prime = sp.diff(f_mvt, x)
c_solution = sp.solve(f_mvt_prime - average_rate_of_change, x)

print(f"\n Mean Value Theorem:")
print(f"  - Average rate of change of f(x) = x^2 on [{a}, {b}] is {average_rate_of_change}")
print(f"  - The value of c such that f'(c) equals the average rate of change is: {c_solution}")

# Define the symbols
x, y = sp.symbols('x y')

# Define the function
f = x**2 + y**2

# Compute the Hessian matrix
hessian_matrix = sp.hessian(f, (x, y))

# Display the Hessian matrix
print("Hessian Matrix:")
print(hessian_matrix)

# Convert to a NumPy array for further numerical computations if needed
hessian_numpy = np.array(hessian_matrix).astype(np.float64)

print("\nHessian Matrix as NumPy Array:")
print(hessian_numpy)
```

# Taylor's Theorem

Taylor's theorem provides a way to approximate a differentiable function around a specific point using polynomials. In particular, the theorem states that for a function $ f $ with $ m $ continuous derivatives, the value of $ f(b) $ can be expressed as a Taylor series expansion around a point $ a $. For $ m = 2 $, the approximation includes the first derivative term and a second-order error term, represented as
$$
f(b) \approx f(a) + (b - a)f'(a) + \frac{1}{2}(b - a)^2f''(\xi)
$$
for some $ \xi $ in the interval between $ a $ and $ b $. In the multivariate case, the theorem generalizes to account for multiple variables, with Taylor series expansions incorporating gradients and Hessians. This provides a powerful tool for approximating functions in higher dimensions and understanding their behavior near specified points. The example illustrates how to apply Taylor's theorem to a specific function, showcasing its utility in both theoretical and practical applications.




# Code demonstration

The provided code implements the third-order Taylor series approximation for a single-variable function. It defines the function $ f(x) = \sin(x) $ and computes its derivatives up to the third order. The code then calculates the Taylor series expansion around the point $ a = 0 $ and constructs the polynomial approximation using these derivatives.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function as sin(x)
def f(x):
    return np.sin(x)  # Example function

#Doing third order approx, so will take derivative three times
def f_prime(x):
    return np.cos(x)  # First derivative

def f_double_prime(x):
    return -np.sin(x)  # Second derivative

def f_triple_prime(x):
    return -np.cos(x)  # Third derivative

# Taylor series approximation at a point a
def taylor_approximation(x, a):
    return (f(a) +
            f_prime(a) * (x - a) +
            (f_double_prime(a) / 2) * (x - a)**2 +
            (f_triple_prime(a) / 6) * (x - a)**3)

# Print the third-order Taylor series polynomial
def print_taylor_polynomial(a):
    # Coefficients
    f_a = f(a)
    f_prime_a = f_prime(a)
    f_double_prime_a = f_double_prime(a) / 2
    f_triple_prime_a = f_triple_prime(a) / 6

    # Want to print the polynomial
    print(f'The third-order Taylor polynomial of f(x) around x = {a} is:')
    print(f'P(x) = {f_a} + {f_prime_a} * (x - {a}) + {f_double_prime_a} * (x - {a})^2 + {f_triple_prime_a} * (x - {a})^3')

# Define the point of expansion and the range of x values
a = 0  # Point of expansion
print_taylor_polynomial(a)

x_values = np.linspace(-2 * np.pi, 2 * np.pi, 400)  # Range of x values
taylor_values = taylor_approximation(x_values, a)
true_values = f(x_values)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(x_values, true_values, label='f(x) = sin(x)', color='blue')
plt.plot(x_values, taylor_values, label='Taylor Approximation (3rd order)', color='orange', linestyle='--')
plt.scatter([a], [f(a)], color='red')  # Point of expansion
plt.title('Taylor Series Approximation of f(x) = sin(x) at x = 0 (3rd Order)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
